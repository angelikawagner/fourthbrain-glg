```python
from pathlib import Path
from gensim import corpora
from gensim.models import LdaModel

import pyLDAvis
import pyLDAvis.gensim_models as gensimvis

# import pandas as pd
```

```python
MODEL_NAME = "lda_50i4p24t" #"lda_50i4p24t"
DATASET = "sample50k_topic4_v2" # "sample50k_health_tech"

DICTIONARY_PATH = Path("../results/dictionary")
CORPUS_PATH = Path("../results/corpus")
MODEL_PATH = Path("../results/models")/DATASET
INFERENCE_PATH = (Path("../results/inference")/DATASET)

INFERENCE_PATH.mkdir(exist_ok=True)
```

```python
# load saved dictionary and corpus
dictionary = corpora.Dictionary.load(f"{DICTIONARY_PATH/DATASET}")
corpus = corpora.MmCorpus(f"{CORPUS_PATH/DATASET}.mm")

# temp_file = datapath()
lda = LdaModel.load(f"{MODEL_PATH}/{MODEL_NAME}.model")
```

```python

vis_data = gensimvis.prepare(lda, corpus, dictionary)
pyLDAvis.display(vis_data)
```

Create mapping of topic numbers between gensim LDAModel and pyLDAVis.

```python
gensim_lavis_map = vis_data.topic_coordinates.reset_index()[["topic", "topics"]] # gensim, pyLDAvis
gensim_lavis_map.rename(columns = {"topic": "gensim_topic", "topics": "pyLDAvis_topic"}, inplace=True)
gensim_lavis_map.to_csv(f"{INFERENCE_PATH}/gensim_pyldavis_map_{MODEL_NAME}.csv", index=False)
```
